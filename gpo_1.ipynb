```python
# last part of the homework
import pandas as pd 
```

```python
data_wt = pd.read_csv('/content/sample_data/wt.csv')
print(data_wt)
```

```python
#In the homework we'll use from scipy.optimize import curve_fit
from scipy.stats import linregress

#Fit a line to the data
slope, intercept, r_value, p_value, std_err = linregress(data_wt['colA'],data_wt['colB'])

#Create a figure instance and an axis for plotting
fig, ax = plt.subplots()

#Plot the line and scatter the data
ax.plot(data_wt['colA'], slope*data_wt['colA']+intercept, color = 'r', label = 'Linear Regression')
ax.scatter(data_wt['colA'], data_wt['colB'], color = 'k', label = 'Data')

#Add a legend
ax.legend()

#Label the axes
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Slope: {:.2f}, Intercept {:.2f}'.format(slope, intercept))
```
